```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
model_name="distilbert-base-uncased-finetuned-sst-2-english"
```

```python
model=AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer=AutoTokenizer.from_pretrained(model_name)
```

```python
seq="Netwoking is Powerful amongst all skills"
res=tokenizer(seq)
```

```python
print(res)
```

```python
token=tokenizer.tokenize(seq)
print(token)
ids=tokenizer.convert_tokens_to_ids(token)
print(ids)
dec_string=tokenizer.decode(ids)
print(dec_string)
```

```python
import torch
import torch.nn.functional as F
```

```python
X_train=[
    "I Love Java it is the best",
    "Java and Javascript are not the same"
]
```

```python
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
res=classifier(X_train)
print(res)
```

```python
batch= tokenizer(X_train, padding=True, truncation=True, max_length=512, return_tensors="pt")
print(batch)
```

```python
with torch.no_grad():
  output=model(**batch)
  print(output)
  predictions =F.softmax(output.logits, dim=1)
  print(predictions)
  labels=torch.argmax(predictions,dim=1)
  print(labels)
```
